```python
import pandas as pd

# Example DataFrame
data = {
    "SYMBOL": ["AAPL", "AAPL", "GOOG", "GOOG", "MSFT", "MSFT"],
    "bullish_marubuzo": [True, False, False, True, False, True],
    "bearish_marubuzo": [False, True, False, False, True, False],
    "doji": [False, False, True, False, False, False],
}

df = pd.DataFrame(data)

# Find patterns where column value is True
def find_patterns(dataframe):
    # Melt the DataFrame to make pattern names a single column
    melted = dataframe.melt(id_vars=["SYMBOL"], var_name="Pattern", value_name="Detected")
    display(melted)
    # Filter rows where Detected is True
    filtered = melted[melted["Detected"]]
    display(filtered)
    
    # Group by SYMBOL and aggregate pattern names
    grouped = filtered.groupby("SYMBOL")["Pattern"].apply(list).reset_index()
    display(grouped)
    grouped.rename(columns={"Pattern": "Detected_Patterns"}, inplace=True)
    
    return grouped

# Apply the function
result = find_patterns(df)

# Display the result
print(result)
```
