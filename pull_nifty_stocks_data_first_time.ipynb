```python
today = datetime.date.today()
previous_year = today.year -1
today_last_year = today.replace(year = previous_year)
today_last_year
```

```python
def pull_stock_data(stock_symbol:str) -> pd.DataFrame:
    today = datetime.date.today()
    previous_year = today.year -1
    today_last_year = today.replace(year = previous_year)
    data = stock_df(symbol=stock_symbol, from_date=today_last_year,
            to_date=today, series="EQ")
    
    return data

pull_stock_data("ABB")
```

```python
nifty_top_500.columns
```

```python
master_data = pd.DataFrame()
for stock_symbol in nifty_top_500.Symbol.unique():
    try:
        #stock = 'TATAMOTORS'
        print(stock_symbol)
        data = pull_stock_data(stock_symbol=stock_symbol)
        master_data = pd.concat([master_data,data],axis=0)
    except:
        pass

master_data
```

```python
# Load environment variables from .env file
load_dotenv()

# Access the database URL
db_url = os.getenv('DB_URL')

# Print to confirm (remove in production for security)
print(f"Database URL: {db_url}")
```

```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

# Define table name
table_name = "nifty_top_500_stocks"

# Database connection string
db_url = db_url

# Create an SQLAlchemy engine
engine = create_engine(db_url)

# Write data to the PostgreSQL table
try:
    master_data.to_sql(table_name, engine, index=False, if_exists='replace')  # 'replace' overwrites table if it exists
    print(f"Data successfully written to table '{table_name}' in PostgreSQL!")
except Exception as e:
    print(f"Error: {e}")
```
